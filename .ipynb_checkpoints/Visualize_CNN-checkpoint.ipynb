```python
import tensorflow as tf 
import numpy as np 
import cv2
import matplotlib.pyplot as plt
import glob
import math
```

```python
cnn_model = './cnn_ckpt/'

face_images_path = 'dataset/testing/face/'
nonface_images_path = 'dataset/testing/nonface/'

height,width = 128,128
img_ext = ['.png','.jpg','jpeg','JPEG','.bmp']
```

```python
def read_image(filepath):
    
    images = []
    labels = []
    
    for f in glob.glob(filepath + '**', recursive=True):
        
        if f[-4:] not in img_ext:
            continue
        
        img = cv2.imread(f, 0)
        img = cv2.resize(img, (height,width))
        img = np.asarray(img, dtype='float32')/255
        img = np.reshape(img, (height,width,1))
        images.append(img)
        label = np.zeros([2])
        label[0] = 1.0
        labels.append(label)
        
    
    return (np.asarray(images, dtype='float32'), np.asarray(labels, dtype='float32'))
```

```python
face_images, face_labels = read_image(face_images_path)
nface_images, nface_labels = read_image(nonface_images_path)
```

```python
tf.reset_default_graph()
sess = tf.InteractiveSession()
checkpoint = tf.train.latest_checkpoint(cnn_model)
graph = tf.get_default_graph()
saver = tf.train.import_meta_graph(checkpoint + '.meta')
saver.restore(sess, checkpoint)
```

```python
layer1_filters = (graph.get_tensor_by_name('Conv_1/weights:0'), graph.get_tensor_by_name('Conv_1/biases:0'))
layer2_filters = (graph.get_tensor_by_name('Conv_2/weights:0'), graph.get_tensor_by_name('Conv_2/biases:0'))
layer3_filters = (graph.get_tensor_by_name('Conv_3/weights:0'), graph.get_tensor_by_name('Conv_3/biases:0'))
layer4_filters = (graph.get_tensor_by_name('Conv_4/weights:0'), graph.get_tensor_by_name('Conv_4/biases:0'))
layer5_filters = (graph.get_tensor_by_name('Conv_5/weights:0'), graph.get_tensor_by_name('Conv_5/biases:0'))
layer6_filters = (graph.get_tensor_by_name('Conv_6/weights:0'), graph.get_tensor_by_name('Conv_6/biases:0'))
```

```python
index = 0
input_img = np.reshape(face_images[index], (height,width))
input_label = np.reshape(face_labels[index], (1,2))
plt.imshow(input_img, cmap='gray')
plt.title(input_label)
plt.show()
```

```python
layer_values = sess.run([(layer1_filters[0], layer1_filters[1]), 
                          (layer2_filters[0], layer2_filters[1]),
                          (layer3_filters[0],layer3_filters[1]),
                          (layer4_filters[0],layer4_filters[1]),
                          (layer5_filters[0],layer5_filters[1]),
                          (layer6_filters[0],layer6_filters[1])], 
                         feed_dict={'input:0':np.reshape(input_img, (1,height,width,1)),
                                                                 'target:0': input_label,
                                                                 'dropout:0':1.0})
```

```python
layer1_filter_w = layer_values[0][0]
layer1_filter_b = layer_values[0][1]

number_of_filters = layer1_filter_w.shape[3]
```

```python
columns = 10
rows = math.ceil(number_of_filters/columns)

fig = plt.figure(figsize=(50,70))
for i in range(0, columns*rows):
    try:
        img = layer1_filter_w[:,:,0,i]
        fig.add_subplot(rows,columns, i+1)
        plt.imshow(img, cmap='gray')
        plt.title("Filter "+str(i+1))
        plt.xticks(())
        plt.yticks(())
    
    except IndexError:
        pass
plt.show()
```
